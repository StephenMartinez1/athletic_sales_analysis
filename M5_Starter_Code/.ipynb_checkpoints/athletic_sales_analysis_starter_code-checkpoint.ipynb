### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.

women_foot_df = combo_sales_df.loc[(combo_sales_df["product"] =="Women's Athletic Footwear")]

# Validate
women_foot_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

groupby_women_foot_sold = women_foot_df.groupby(['retailer', 'region', 'state', 'city']).sum('units_sold').reset_index().rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})

# Show the top 5 results.

groupby_women_foot_sold_top5 = groupby_women_foot_sold.sort_values(by=['Womens_Footwear_Units_Sold'], ascending=False, ignore_index=(True))

groupby_women_foot_sold_top5.head()[['retailer', 'region', 'state', 'city', 'Womens_Footwear_Units_Sold']]
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

pivot_women_foot_sold = women_foot_df.pivot_table(index = ['retailer', 'region', 'state', 'city'],
                                                         values = 'units_sold',
                                                         aggfunc = 'sum').reset_index().rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})

# Show the top 5 results.

pivot_women_foot_sold_top5 = pivot_women_foot_sold.sort_values(by=['Womens_Footwear_Units_Sold'], ascending=False, ignore_index=(True))

pivot_women_foot_sold_top5.head()[['retailer', 'region', 'state', 'city', 'Womens_Footwear_Units_Sold']]
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"

pivot_day_sales = women_foot_df.pivot_table(index = 'invoice_date' ,
                                            values = 'total_sales',
                                            aggfunc = 'sum').reset_index().rename(columns={'total_sales':'Total Sales'})

# Show the table.

pivot_day_sales_sorted = pivot_day_sales.sort_values(by='Total Sales', ascending=False)

pivot_day_sales_sorted.head(1)

```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

pivot_day_sales['invoice_date'] = pd.to_datetime(pivot_day_sales['invoice_date'])

pivot_day_sales.set_index('invoice_date', inplace=True)

daily_sales = pivot_day_sales.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".

pivot_day_sales.sort_values("Total Sales", ascending = False)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

pivot_weekly_sales = pivot_day_sales.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".

weekly_sales = pivot_weekly_sales.sort_values(by='Total Sales', ascending=False)

weekly_sales
```
